# Making changes to the data

```python
# create a column

df['Total'] = df['HP'] + df['Attack'] + df['Defense'] + df['Sp. Atk'] + df['Sp. Def'] + df['Speed']

df.head()
```

```python
# create and drop columns with aggregate functions
# sum
df['sumtotal'] = df.iloc[:,4:10].sum(axis=1)

# mean
df['mean'] = df.iloc[:,4:10].mean(axis=1)

# median
df['median'] = df.iloc[:,4:10].median(axis=1)

# std
df['std'] = df.iloc[:,4:10].std(axis=1)

df.head()
```

```python
# drop columns
df = df.drop(columns=['sumtotal', 'mean', 'median', 'std'])

df.head()
```

# Filtering Data

```python
# filter loc
# 1 filter
df.loc[df['Type 1'] == 'Grass']
```

```python
 # 2 filters
df.loc[(df['Type 1'] == 'Grass')& (df['Type 2'] =='Poison')]
```

```python
# 4 filters
df.loc[
    (df['HP'] > 70 )  # you should put a '&' to add another condition
    & (df['Attack'] > 100) 
    & (df['Defense'] > 100) 
    & (df['Speed'] > 90)]
```

```python
# Filter AND/OR
# AND
df.loc[
        (df['HP'] > 110)
         & # '&' is the sign to 'and'
         (df['Attack'] > 120) ]
```

```python
# Filter AND/OR
# OR
df.loc[
        (df['HP'] > 120) 
           | # '|' is the sign to 'or'
        (df['HP'])
      ]
```

# Reset index


```python
# reset index 
new_df = df.loc[(df['HP'] > 110) & (df['Attack'] > 120) ]

new_df = new_df.reset_index()

new_df
```

```python
# reset index and drop old index

new_df2 = df.loc[(df['HP'] > 110) &(df['Attack'] > 120) ]

new_df2 = new_df2.reset_index(drop=True)

new_df2
```

```python
# Filter by substring
# the '~' sign show the 'not in' parameters to string

df.loc[(~df['Name'].str.contains('nder')) | (~df['Name'].str.contains('saur'))]
```
